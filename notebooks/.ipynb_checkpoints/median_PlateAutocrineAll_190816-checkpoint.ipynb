```python
##Functions needed here.
def insertLibrary (row):
    if row['Sample'] == "STE6":
        return("ScSte6")
    if row['Sample'] == "E05973":
        return("YlSte6")
    else:
        return(row['Sample'].split("/")[0])
    
def insertLibraryVer (row):
    if row['Sample'] == "STE6":
        return("")
    if row['Sample'] == "E05973":
        return("")
    else:
        return(row['Sample'].split("/")[1])
    
def insertMnConc (row):
    if row['Sample'] == "STE6":
        return("")
    if row['Sample'] == "E05973":
        return("")
    else:
        return(row['Sample'].split("/")[2])
    
def insertLibraryMisc (row):
    if row['Sample'] == "STE6":
        return("")
    if row['Sample'] == "E05973":
        return("")
    else:
        return(row['Sample'].split("/")[3])
    
def insertGFPdiff (row):
    #if row['oldGFP'].notna() and row['retrafoGFP'].notna():
    return((row['oldGFP']-row['retrafoGFP'])/row['oldGFP'])

def insertRFPdiff (row):
    #if row['oldRFP'].notna() and row['retrafoRFP'].notna():
    return((row['oldRFP']-row['retrafoRFP'])/row['oldRFP'])

def insertGFPRFPratio (row):
    #if row['oldRFP'].notna() and row['retrafoRFP'].notna():
    return((row['retrafoGFP']/row['retrafoRFP']))
```

### Explanation of correlations
I think I have figured out the issue. There is a correlation between the GFP and RFP measurements between the re-trafo and old clones. But there is no correlation between (RFP and GFP) within the re-trafo clones and old clones. However it looks like the (oldGFP - retrafoGFP) and (oldRFP - retrafoRFP) variables are correlated. I was confused about this, but I think the answer is that in general the selection pushed increased signal in both GFP and RFP making it look like there is a correlation.  
The real answer is by looking at the (GFPdiff and oldGFP) which theoretically should be positive since better clones should get proportionally better with higher expression (which we have already established). But looking at this it is very clear there is no correlation at all... Everything makes sense now?
